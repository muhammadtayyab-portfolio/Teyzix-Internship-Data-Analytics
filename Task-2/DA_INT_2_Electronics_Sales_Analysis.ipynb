## 2. Dataset Generation (Faker)
Generate a realistic synthetic dataset (1200+ records) simulating an electronics store's sales across UAE cities.

```python
# Categories, products, and price ranges (AED)
category_products = {
    "Mobiles": ["iPhone 15", "Samsung Galaxy S24", "Xiaomi Redmi Note 13", "Google Pixel 8", "OnePlus 12"],
    "Laptops": ["MacBook Air M3", "Dell XPS 13", "HP Pavilion 15", "Lenovo ThinkPad E14", "Asus ZenBook 14"],
    "Tablets": ["iPad Air", "Samsung Galaxy Tab S9", "Lenovo Tab P11", "Xiaomi Pad 6"],
    "Audio": ["Sony WH-1000XM5", "AirPods Pro 2", "JBL Flip 6", "Bose QuietComfort"],
    "TVs": ["Samsung 55in QLED", "LG 65in OLED", "Sony 50in Bravia", "Hisense 43in 4K"],
    "Gaming": ["PlayStation 5", "Xbox Series X", "Nintendo Switch OLED", "Logitech G502 Mouse"],
    "Smart Home": ["Amazon Echo Dot", "Google Nest Hub", "Philips Hue Starter Kit", "Ring Video Doorbell"],
    "Accessories": ["Anker Power Bank", "USB-C Cable Pack", "Wireless Charger Pad", "Laptop Sleeve"],
    "Wearables": ["Apple Watch Series 9", "Samsung Galaxy Watch 6", "Fitbit Charge 6", "Garmin Venu 3"],
    "Cameras": ["Canon EOS R50", "Sony Alpha A6400", "GoPro Hero 12", "DJI Osmo Pocket 3"]
}

category_price_range = {
    "Mobiles": (800, 4500), "Laptops": (1500, 9000), "Tablets": (700, 3500),
    "Audio": (150, 1800), "TVs": (1200, 8000), "Gaming": (300, 2500),
    "Smart Home": (150, 1200), "Accessories": (30, 350),
    "Wearables": (300, 2800), "Cameras": (1000, 6500)
}

uae_cities = ["Dubai", "Sharjah", "Abu Dhabi", "Ajman", "Al Ain", "Ras Al Khaimah", "Fujairah"]
payment_methods = ["Credit Card", "Debit Card", "Cash", "Apple Pay", "Tabby (BNPL)"]
store_branches = ["Mall Branch", "Downtown Branch", "Online Store", "Outlet Branch"]
```

```python
# Customer pool - keeps age/gender/city consistent per customer across repeat purchases
NUM_CUSTOMERS = 280
customers = {}
for i in range(1, NUM_CUSTOMERS + 1):
    cust_id = f"CUST{i:04d}"
    customers[cust_id] = {
        "age": random.randint(18, 65),
        "gender": random.choice(["Male", "Female"]),
        "city": random.choice(uae_cities)
    }
customer_ids = list(customers.keys())
```

```python
# Weighted random date - higher density in DSF/Ramadan/Eid/Black Friday months
def random_purchase_date():
    month_weights = {1: 3, 2: 1, 3: 2, 4: 1, 5: 1, 6: 2, 7: 1, 8: 1, 9: 1, 10: 1, 11: 3, 12: 3}
    month = random.choices(list(month_weights.keys()), weights=list(month_weights.values()))[0]
    day = random.randint(1, 28)
    return datetime(2025, month, day)
```

```python
# Main generation loop
NUM_ROWS = 1200
records = []

for i in range(1, NUM_ROWS + 1):
    txn_id = f"TXN{i:05d}"
    cust_id = random.choice(customer_ids)
    cust_info = customers[cust_id]

    category = random.choice(list(category_products.keys()))
    product = random.choice(category_products[category])
    min_p, max_p = category_price_range[category]
    unit_price = round(random.uniform(min_p, max_p), 2)

    quantity = random.choices([1, 2, 3, 4, 5], weights=[60, 20, 10, 5, 5])[0]
    discount_pct = random.choices([0, 5, 10, 15, 20], weights=[60, 15, 15, 5, 5])[0]
    total_amount = round(quantity * unit_price * (1 - discount_pct / 100), 2)
    purchase_date = random_purchase_date()

    records.append({
        "Transaction_ID": txn_id, "Customer_ID": cust_id,
        "Customer_Age": cust_info["age"], "Customer_Gender": cust_info["gender"],
        "Product_Category": category, "Product_Name": product,
        "Quantity": quantity, "Unit_Price": unit_price, "Discount_Percent": discount_pct,
        "Total_Amount": total_amount, "Purchase_Date": purchase_date.strftime("%Y-%m-%d"),
        "Customer_City": cust_info["city"], "Payment_Method": random.choice(payment_methods),
        "Store_Branch": random.choice(store_branches),
        "Rating": random.choices([1,2,3,4,5], weights=[3,5,15,37,40])[0]
    })

df = pd.DataFrame(records)
print(df.shape)
df.head()
```
